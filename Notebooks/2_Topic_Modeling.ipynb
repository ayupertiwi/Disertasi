## Topic Modeling

```python
stopwords = []
f = open('/content/data/stopwords.txt', 'r')
for i in f.readlines():
    stopwords.append(i.replace('\n', ''))
sw_eng = stopwords + list(punctuation)
```

```python
from src.topic_detection import DynamicTopicModeling, CustomLatentDirichletAllocation
```

```python
# parameter yg dicari
search_params = {'n_components': [8],
                'learning_decay': [0.5]}

# model lda
lda = CustomLatentDirichletAllocation(learning_method='online', max_iter=25, random_state=42)

# bow
bow = CountVectorizer(tokenizer=word_tokenize, min_df=5, stop_words=sw_eng)
```

## Topic Modeling LDA

```python
model_dtm = DynamicTopicModeling(df, lda, bow, 'abstract_clean', 'year')
best_model_params = model_dtm.search_number_topic(search_params)
model_dtm.training(best_model_params, topk=35)
```

Using grid search is an automatic option, if you want to find hyperparameters with the best perplexity score

```python
model_dtm.training_by_periods(periods=3, best_params=best_model_params)
```
