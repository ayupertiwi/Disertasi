## Cleaning

```python
import nltk
nltk.download('wordnet')
```

```python
def cleansing(text):
    lemmatizer = WordNetLemmatizer()
    text = str(text).lower().strip()
    word_list = word_tokenize(text)
    word_list = [lemmatizer.lemmatize(word) for word in word_list]
    word_list = [word for word in word_list if word.isalpha() and len(word)>2]
    text = ' '.join(word_list)
    return text
```

```python
df.abstract[0]
```

```python
df['abstract_clean'] = df.abstract.apply(cleansing)
```

```python
df.abstract_clean[0]
```

## Create Stopwords

```python
bow = CountVectorizer()
bow_matrix = bow.fit_transform(df.abstract_clean)
```

```python
term = bow.get_feature_names_out()

sums = bow_matrix.sum(axis=0)

data = []

for col, term in enumerate(term):
    data.append((term, sums[0, col]))

ranking = pd.DataFrame(data, columns=['term', 'frekuensi'])
ranking.head()
```

```python
stopwords = ranking.sort_values('frekuensi', ascending=False).head(200)
stopwords
```
